```python
import pandas as pd
import numpy as np

transition0713 = pd.read_stata("listall0713.dta")
transition1317 = pd.read_stata("listall1317.dta")
```

```python
transition0713.gez3
```

```python
transition0713 = transition0713[["index", "DID","debt", "education", "gender", "health", "gez3", "netWorth", "personid", "ntot", "time", "treated", "unemployed", "wealth"]]
transition1317 = transition1317[["index", "DID","debt", "education", "gender", "health", "gez3", "netWorth", "personid", "ntot", "time", "treated", "unemployed", "wealth"]]
```

```python
def selfAssessedHealth(row):
    healthValues = {"Excellent" : 5, "Good" : 4, "Fair" : 3, "Not so good" : 2, "Poor" : 1,
                    "excellent" : 5, "good" : 4, "fair" : 3, "not so good" : 2, "poor" : 1} 
    return healthValues[row["gez3"]]
```

```python
transition0713["health"] = transition0713.apply(selfAssessedHealth, axis = 1)
transition1317["healht"] = transition1317.apply(selfAssessedHealth, axis = 1)
```

```python
listStart0713 = transition0713.loc[transition0713["time"] == 0]
listStop0713 = transition0713.loc[transition0713["time"] == 1]
listStart1317 = transition1317.loc[transition1317["time"] == 0]
listStop1317 = transition1317.loc[transition1317["time"] == 1]
```

```python
list(listStart0713)
```

```python
def startWealth(row):
    print(row["netWorth"])
    return row["netWorth"]
```

```python
listStart0713["startWealth"] = listStart0713.apply(startWealth, axis = 1)
listStart1317["startWealth"] = listStart1317.apply(startWealth, axis = 1)
```

```python
merged0713 = pd.merge(listStart0713, listStop0713, on = "personid")
merged1317 = pd.merge(listStart1317, listStop1317, on = "personid")
```

```python
def healthTransitionPos(row):
#     print(row.health_x)
    if row.health_x < row.health_y:
        print("ja")
        return 1
    else:
        return 0
```

```python
merged0713["transitioned"] = merged0713.apply(healthTransition, axis = 1)
merged1317["transitioned"] = merged1317.apply(healthTransition, axis = 1)
```
