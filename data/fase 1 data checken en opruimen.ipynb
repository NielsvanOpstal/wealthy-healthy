```python
# Import the data
import pandas as pd
dataAgw = pd.read_stata("agw2017en_1.0.dta")
```

```python
# Eerste check voor rows met een of meer NaN waarde.
nNans = 0
for x in range(0,dataAgw.shape[0]):
    
    if dataAgw.loc[x].isnull().any():
        nNans += 1
# Print het totaal aantal rows met een NaN waarde
print("totaal aantal NaN rows: " + str(nNans))
```

 Blijkt dus te zijn dat er twee soorten NaN rows zijn. De eerste, en bruikbare, noem ik "hypotheek NaNs". Dat zijn rows die enkel NaN waarden hebben bij de huizenbezit componenten (dus alles vanaf b26oga). Deze NaNs zullen geen probleem moeten opleveren omdat de rest van de componenten wel goed is ingevuld.
De tweede en onbruikbare soort NaN zijn rows waarbij zo goed als alle of alle waarden NaN zijn. Deze zal ik eruit filteren


```python
# Hier check ik hoeveel onbruikbare Nans er zijn, oftewel rows met Nans voor de component B260ga
nONans = 0
for x in range(0,dataAgw.shape[0]):
    # hier check ik of er NaN waarden zijn in de column voor de 
    if dataAgw.loc[x,:"x1c"].isnull().any():
        # Hier uit blijkt dat alle onbruikbare NaN rows bij b1b ook een NaN hebben, waarop ik dus kan filteren
        print(dataAgw.loc[x,"b1b"])
        nONans += 1
# Print het totaal aantal rows met een NaN waarde
print("aantal onbruikbare rows:" + str(nONans))
```

```python
# Alle onbruikbare NaN rows hadden dus in gemeen dat alle waarden vanaf b1a tot in iedergeval b26oga NaN waren. 
# Hier filter ik deze rows eruit. Zo blijft er dus een dataset over waar geen onbruibare waardes in zitten.
agwFiltered = dataAgw[dataAgw.b1b.notnull()]
```
